```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
```

```python
df = pd.read_csv("../data/raw/creditcard.csv")
```

```python
df.describe()
```

```python
fraud_transactions   = df[df['Class'] == 1]
genuine_transactions = df[df['Class'] == 0]

print("Total number of record              : {}".format(df.shape[0]))
print("Total number of features in dataset : {}".format(df.shape[1]))
print("Total number of positive cases      : {}".format(fraud_transactions.shape[0]))
print("Percentage of positive cases        : {}".format(100 * df[df['Class'] == 1].shape[0]/df.shape[0]))
print("Outlier fraction                    : {}".format(len(fraud_transactions)/len(genuine_transactions)))
fig = plt.figure(figsize = (10,12))
sns.countplot(df["Class"])
```

```python
#Function to generate a random color
def getRandomColor():
    r = rnd.randint(0,255)
    g = rnd.randint(0,255)
    b = rnd.randint(0,255)
    print((r,g,b))
    return (r,g,b)

def plotHistograms(df, features):
    for feature in features:
        f, axis = plt.subplots(1,1, figsize = (9,4))
        input_distribution_data = df[feature]
        sns.histplot(input_distribution_data, ax = axis,  kde = True)
```

```python
plotHistograms(df, df.columns)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
scaler = MinMaxScaler()
X = df.drop(['Class'], axis =1)
X = scaler.fit_transform(X.to_numpy())

pca = PCA(n_components = 2)
pca_features = pca.fit_transform(X)

```

```python
genuine_cases = pca_features[df[df['Class'] == 0].index]
fraud_cases = pca_features[df[df['Class'] == 1].index]
```

```python
plt.scatter(genuine_cases[:,0],genuine_cases[:,1],color='red')
plt.scatter(fraud_cases[:,0],fraud_cases[:,1],color='blue')
plt.show()
```
